# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def view(image, name_of_window="window"):
    cv2.namedWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
img = cv2.imread(cv2.samples.findFile("corona.png"))
hsv= cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
b, g, r= cv2.split(img)
h, s, v = cv2.split(hsv)

colors= ['blue', 'green', 'red']

for i, color in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [255], [0, 255])
    plt.plot(hist, color= color)
    plt.title("BGR in the Image")
    plt.show
```

```python
hist_h= cv2.calcHist([img], [0], None, [255], [0,255])
plt.plot(hist_h)
plt.title("H in the Image")
plt.show
```

```python
colors= ['blue', 'green', 'red']
#h-blue , green- s, red- v
for i, color in enumerate(colors):
    hist = cv2.calcHist([hsv], [i], None, [255], [0, 255])
    plt.plot(hist, color= color)
    plt.title("HSV in the Image")
    plt.show
```
